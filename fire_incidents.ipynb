# Top 20 neighborhood with the most fire incidents

```python

import pandas as pd


file = pd.read_csv('fireIncident.csv')


neighborhoodCount = file['neighborhood'].value_counts().sort_values().head(50)

print(f"Top {neighborhoodCount}:")
    




```

```python

import matplotlib.pyplot as plt


file = pd.read_csv('fireIncident.csv')


neighborhood_counts = file.groupby('neighborhood').size()

plt.figure(figsize=(12, 8))
neighborhood_counts.plot(kind='bar', color='blue')


plt.xlabel('Neighborhood')
plt.ylabel('Count of fire')
plt.title('Number of fire by Neighborhood')

plt.tight_layout()

plt.show()
```
