# What is the Cleanest Neighborhood in Pittsburgh?

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

## Sub-metric 1: Trees (Riley)

```python
trees = pd.read_csv("trees.csv")
tree_counts = trees["neighborhood"].value_counts().sort_values(ascending=False)

plt.xlabel("Neighborhood")
plt.ylabel("Trees")

tree_counts[:10].plot(kind='bar')
```

## Sub-metric 2: Fires (Henry)

```python
fires = pd.read_csv('fireIncident.csv')
fire_counts = fires['neighborhood'].value_counts().sort_values()
```

```python
plt.figure(figsize=(12, 8))
fire_counts[:10].plot(kind='bar', color='blue')

plt.xlabel('Neighborhood')
plt.ylabel('Fire Counts')
plt.title('Fire Counts by Neighborhood')

plt.tight_layout()

plt.show()
```

## Sub-metric 3: Trash Containers (Riley)

```python
trash = pd.read_csv("smart_trash_containers.csv")
trash_counts = trash["neighborhood"].value_counts().sort_values(ascending=False)

trash_counts[:10].plot(kind='bar')

plt.xlabel("Neighborhood")
plt.ylabel("Trash Containers")

plt.show()
```

## Putting it all Together

```python
average = {}
for key in trash_counts.index:
    average[key] = 0
    for count in [trash_counts, tree_counts, fire_counts]:
        average[key] += (count.index.get_loc(key)+1)
    average[key] /= 3
sorted_average = sorted(average.items(), key=lambda x:x[1])
```
